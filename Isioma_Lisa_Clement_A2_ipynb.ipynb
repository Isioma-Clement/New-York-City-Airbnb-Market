```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
```

```python
# import dataset

air_bnb = pd.read_csv('Airbnb_NYC_2019.csv', index_col = 0)
air_bnb
```

```python
# create dataframe for airbnbs in Manhattan alone

Manhattan = air_bnb.loc[['Manhattan' in neighbourhood for neighbourhood in air_bnb.neighbourhood_group],:]
Manhattan
```

```python
# show distribution of airbnbs in Manhattan by price

plt.hist(Manhattan['price'], bins = np.arange(0,1700, 50))
plt.xlabel('Prices in USD')
plt.ylabel('count')
plt.title('Distribution of prices of Airbnb listings in Manhattan, 2019')
plt.show()
```

```python
# use scatterplot to show relationship between number of reviews and prices

plt.scatter(Manhattan['price'],Manhattan['number_of_reviews'],alpha= 0.5, s=7)
plt.xlabel('Price(USD)')
plt.ylabel('Number of Reviews')
plt.title('Scatter Plot of Number of Reviews vs Price')
plt.xlim(0, 3000)
plt.show
```

```python
# Get most popular neighbourhoods in Manhattan for airbnb rentals
# identify the top 10 neighbourhoods in Manhattan by number of listing

top_manhattan=(air_bnb.loc[['Manhattan' in neighbourhood for neighbourhood in air_bnb.neighbourhood_group],:]).groupby('neighbourhood').sum().sort_values('calculated_host_listings_count', ascending = 0).iloc[0:10]
top_manhattan.reset_index()
```

```python
top_manhattan.shape
```

```python
#create a subset data of the popular neighbourhoods

top_10 = Manhattan[ ((Manhattan['neighbourhood'] == 'Financial District') | (Manhattan['neighbourhood'] == 'Upper West Side') | (Manhattan['neighbourhood'] == 'Hell\'s Kitchen') | (Manhattan['neighbourhood'] == 'Murray Hill') | (Manhattan['neighbourhood'] == 'Upper East Side') | (Manhattan['neighbourhood'] == 'Midtown') | (Manhattan['neighbourhood'] == 'Theater District') | (Manhattan['neighbourhood'] == 'Chelsea') | (Manhattan['neighbourhood'] == 'Tribeca') | (Manhattan['neighbourhood'] == 'Kips Bay') )]
top_10
```

```python
import seaborn as sn
```

```python
# count plot to visualize how many Airbnb listings are in each of those “top 10” neighbourhoods in Manhattan in 2019

sn.countplot(x='neighbourhood', data = top_10)
plt.title("Airbnb Listing Count Per Top Neighbourhood in Manhattan, 2019")
plt.xticks(rotation=45, horizontalalignment='right')
plt.show()
```

```python
# resolve overlapping

plt.figure(figsize=(18, 9))
sn.countplot(x='neighbourhood', data = Manhattan)
plt.title("Airbnb Listing Count Per Top Neighbourhood in Manhattan,2019")
plt.xticks(rotation=45, horizontalalignment='right')
plt.show()
```
